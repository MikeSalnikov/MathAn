```python
import numpy as np
```

```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

**Задача 2**

Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

1) Применим формулу Пуассона, зная, что $m = 0$ и $λ = np = 5000 \cdot 0.0004 = 2$:

$$P_{0} \approx \frac{\lambda^m}{m!}e^{-\lambda} \approx \frac{2^{0}}{2!}e^{-2} \approx  0.135$$

```python
n = 5000
m = 0
p = 0.0004
lambda_ = n * p

(lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)
```

2) Перегорит 2 лампочки
Применим формулу Пуассона, зная, что $m = 2$ и $λ = np = 5000 \cdot 0.0004 = 2$:

$$P_{2} \approx \frac{\lambda^m}{m!}e^{-\lambda} \approx \frac{2^{2}}{2!}e^{-2} \approx  0.2706$$

```python
n = 5000
m = 2
p = 0.0004
lambda_ = n * p

(lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)
```
